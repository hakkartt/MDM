```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('communties.csv')[['Id', 'modularity_class']].astype({'Id': int, 'modularity_class': int}).set_index('Id')
data
```

```python
kf = pd.read_csv('schoolclass5meta.txt', sep='\t')
df = pd.DataFrame(columns=['Id', 'class', 'gender'])
c = kf.columns
df.at[0, :] = kf.columns
kf = kf.rename(columns={c[0]: 'Id',
                        c[1]: 'class',
                        c[2]: 'gender'
})
metadata = pd.concat([df, kf], axis=0).astype({'Id': int, 'class': str, 'gender': str}).set_index('Id')
print(metadata)
```

```python
df = pd.concat([metadata, data], axis=1)
print(df[df.modularity_class == 0.0])
```
